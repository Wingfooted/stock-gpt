```python
class DB:
    def __init__(self):
        self.user = "root"
        self.port = 8000
        self.db = "earnings"
        self.connection = pymysql.connect(host="127.0.0.1",
                                     port=self.port,
                                     user=self.user,
                                     db=self.db)

    def __call__(self, query):
        with self.connection.cursor() as cursor:
            cursor.execute(query)
            data = cursor.fetchall()
        return data
    
db = DB()

bsa = set(db("SELECT DISTINCT act_symbol FROM balance_sheet_assets"))
cf = set(db("SELECT DISTINCT act_symbol FROM cash_flow_statement"))
ist = set(db("SELECT DISTINCT act_symbol FROM income_statement"))
eps = set(db("SELECT DISTINCT act_symbol FROM eps_estimate"))

# Find the intersection of all three sets
described = bsa.intersection(cf, ist, eps)

# Output the result
print("Number of fully described stocks:", len(described))
#print("Fully described stocks:", described)

blacklist = bsa.union(cf, ist, eps) - described
print("Number of stocks with incomplete data:", len(blacklist))

symbols = [s[0] for s in described]
```

```python
#longcell
from tqdm import tqdm
tickers = yf.Tickers(' '.join(symbols))
```

```python
# Initialize lists to store data
symbol_list = []
country_list = []
sector_list = []
industry_list = []
len_business_sum_list = []
currency_list = []

# Example loop fetching data (replace with actual data retrieval)
for s in tqdm(symbols):
    info = tickers.tickers[s].info  # Replace with your method of fetching info
    symbol_list.append(s)
    country_list.append(info.get('country', ''))
    sector_list.append(info.get('sector', ''))
    industry_list.append(info.get('industry', ''))
    len_business_sum_list.append(len(info.get('longBusinessSummary', '')))
    currency_list.append(info.get('currency', ''))

# Create DataFrame
df = pd.DataFrame({
    'Country': country_list,
    'Sector': sector_list,
    'Industry': industry_list,
    'Length_of_Business_Summary': len_business_sum_list,
    'Currency': currency_list
}, index=symbol_list)

# Display the DataFrame
print(df)
```

```python
def encode_categorical_columns(df):
    # Exclude 'Symbol' and 'Length_of_Business_Summary' columns
    columns_to_encode = [col for col in df.columns if col != 'Symbol' and col != 'Length_of_Business_Summary']
    
    # Perform one-hot encoding on selected columns
    encoder = OneHotEncoder(sparse_output=False)
    encoded_features = encoder.fit_transform(df[columns_to_encode])
    
    # Get feature names for the encoded columns
    encoded_column_names = encoder.get_feature_names_out(columns_to_encode)
    
    # Create DataFrame with encoded columns
    df_encoded = pd.DataFrame(encoded_features, columns=encoded_column_names, index=df.index)
    
    # Concatenate original 'Symbol' column with encoded DataFrame
    
    return df_encoded
df_encoded = None
df_encoded = encode_categorical_columns(df)
df_encoded
```
